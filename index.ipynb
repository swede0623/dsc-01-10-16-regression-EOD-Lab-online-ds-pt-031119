### Let's do Regression 

Right here is the real deal. Let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). We shall do this is a loop and in every iteration, we shall pick one of the independent variables  perform following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`.
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration:
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats
```

```python
data.columns
```

```python
ind_var_columns = data.drop('medv', axis=1).columns
ind_var_columns
```

```python
tar_v = 'medv'
model_values = []
for idv in ind_var_columns:
    
    #step 1 - run regression
    f = tar_v+"~"+idv
    model = ols(formula=f, data=data).fit()
    print(f)
    
    #step 2 - plot data and regression line
    plt.scatter(df[idv], df[tar_v])
    plt.plot(df[idv], model.predict(df[idv]), c='red')
    plt.title(idv)
    plt.show()
    
    #plot residuals
    fig = plt.figure(figsize=(15,8))
    fig = sm.graphics.plot_regress_exog(model, idv, fig=fig)
    plt.show()
    
    #plot QQ plot
    resid = model.resid
    fig2 = sm.graphics.qqplot(resid, dist=stats.norm, line='45', fit=True)
    plt.show()
    
    #store model values in array
    jb = sm.stats.jarque_bera(resid)
    model_values.append([idv, model.rsquared, model.params[0], model.params[1], model.f_pvalue, jb[0]])
```

```python
regression_stats = pd.DataFrame(model_values)
```

```python
regression_stats.columns = ['ind_var', 'r_squared', 'intercept', 'slope', 'p_value', 'normality_JB']
```
